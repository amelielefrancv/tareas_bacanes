```python
#Formalidad Titulo Problema
print("Problema 2: Programa que imprima los números del -100 al 200 que sean divisibles entre 6")

#Codigo para generar lista de numeros desde el -100 hasta el 200 (debemos escribir 201)
for x in range(-100,201):

#Condición: Formula para que los numeros divisibles en 6
  if x % 6 == 0:

#Si la condición se cumple, se imprime el numero
      print(x)
```

```python
#Formalidad Titulo Problema
print("Problema 3: Pseudocódigo, Sumador y comparador de dos números")

#Datos: Usuario ingresa dos numeros
numero1 = int(input("Ingrese el primer numero:"))
numero2 = int(input("Ingresa el segundo numero:"))

#Suma: Se suman los dos numeros colocados
suma = numero1 + numero2

# Condición: Si la suma es menor a 200
if suma < 200:
    resultado = "El resultado es menor a 200"
else:
    # Condición: Si la suma está entre 200 y 249
    if suma > 200 and suma < 250:
        resultado = "El resultado es mayor a 200 y menor a 250"
    else:
        # Condición: Si la suma es mayor a 250 o igual a 200 o 250
        resultado = "El resultado es mayor a 250"


#Imprime resultado de la suma y mensaje
print("La suma es:", suma)
print(resultado)

```

```python
#Formalidad Titulo Problema
print("Problema 4: Pseudocódigo, Calculador de promedio")

#El usuario ingresa su edad
edad = int(input("Ingresa tu edad"))

#Le preguntamos si le gusta o no la musica urbana
respuesta = input("¿Te gusta la musica urbana? (si/no)")

#Codigo para convertir la respuesta "si" en boolean_verdad
gusta_musica_urbana = respuesta == "si"

#Se evaluan las combinaciones posibles de respuestas (edad/gustos)
if edad >= 18 and gusta_musica_urbana:
    resultado = "Eres mayor de edad y te gusta la musica urbana"
else:
    if edad >= 18 and not gusta_musica_urbana:
        resultado = "Eres mayor de edad pero no te gusta la música urbana"
    else:
        if edad < 18 and gusta_musica_urbana:
            resultado = "Eres menor de edad y te gusta la música urbana"
        else:
            resultado = "No eres mayor de edad y no te gusta la musica urbana"

#Se imprime el resultado final
print(resultado)
```
